```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Загрузка данных
data = load_iris()
X = data.data  # Входные данные
y = data.target  # Метки классов

# ====== Метод PCA ======

# Создание объекта PCA и уменьшение размерности до 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Визуализация результатов PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="viridis", s=50)
plt.title("PCA (2 компоненты)")
plt.xlabel("Главная компонента 1")
plt.ylabel("Главная компонента 2")
plt.colorbar()
plt.show()

# ====== Метод t-SNE ======

# Создание объекта t-SNE и уменьшение размерности до 2D
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Визуализация результатов t-SNE
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap="viridis", s=50)
plt.title("t-SNE (2 компоненты)")
plt.xlabel("Компонента 1")
plt.ylabel("Компонента 2")
plt.colorbar()
plt.show()
```
