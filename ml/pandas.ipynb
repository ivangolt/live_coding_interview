Сделать в таблице колонку со средней суммой продаж для клиента за три дня:
день в записи, предыдущий и следующий.

```python
import pandas as pd

# Входные данные
data = {
    "client_id": [1, 1, 1, 2, 2, 2, 3, 3, 3],
    "sum": [30, 12, 6, 30, 15, 120, 0, 3, 12],
    "day": [
        "2020-04-01",
        "2020-04-03",
        "2020-04-04",
        "2020-04-02",
        "2020-04-03",
        "2020-04-04",
        "2020-04-03",
        "2020-04-04",
        "2020-04-05",
    ],
}

# Преобразование в DataFrame
df = pd.DataFrame(data)
df
```

```python
df["day"] = df["day"].astype("datetime64[ns]")
df
```

```python
df = df.sort_values(by=["client_id", "day"])
df
```

```python
df["mean_3_days"] = (
    df.groupby("client_id")["sum"]
    .rolling(window=3, min_periods=1)
    .mean()
    .reset_index(0, drop=True)
)
df
```
